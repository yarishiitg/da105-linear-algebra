```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)

# Plot with format string
plt.plot(x, y, 'ro')  # Red circles

# Plot with additional keyword arguments
plt.plot(x, y, label='Data 1', linewidth=2, color='blue')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Plot')
plt.legend()
plt.show()
```
